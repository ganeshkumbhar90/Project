### Removing the skewness using log transformation

```python
df.head()
```

```python
df['W']=np.log(df['W'])
df['R']=np.log(df['R'])
df['AB']=np.log(df['AB'])
df['H']=np.log(df['H'])
df['2B']=np.log(df['2B'])
df['3B']=np.log(df['3B'])
df['HR']=np.log(df['HR'])
df['BB']=np.log(df['BB'])
df['SO']=np.log(df['SO'])
df['SB']=np.log(df['SB'])
df['RA']=np.log(df['RA'])
df['ER']=np.log(df['ER'])
df['ERA']=np.log(df['ERA'])
df['CG']=np.log(df['CG'])
df['SHO']=np.log(df['SHO'])
df['SV']=np.log(df['SV'])
df['E']=np.log(df['E'])
```

```python
# Validating the skewness
sk=df.skew()
sk[np.abs(sk)>0.5].all()
```

As we see the above showing true so we can say that skewness is removed within range(-0.5-0.5)

## Checking correlation for multicolinearity using VIF

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
# finding vif in each scaled column
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif=pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i)for i in range(x_scaled.shape[1])]
vif['features']=x.columns
vif
```

```python
# Dropping ER column and columns R ab and h as they dont have any corelaltion
x.drop(columns=['ER','R','AB','H'],axis=1,inplace=True)
```

## Split Data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.25,random_state=1352)
```

## Model instansiation and trainig

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

### Adjusted r2 square

```python
# For training data
regression.score(x_train,y_train)
```

```python
# For testing data
regression.score(x_test,y_test)
```

```python
y_pred=regression.predict(x_test)
```
